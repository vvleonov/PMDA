##### 1. Построить математическую модель и численно решить задачу:
*Колония простейших растёт с постоянной удельной интенсивностью роста 0,7 на одну особь в день. Изначально колония состоит из двух особей. Найти размер колонии через шесть дней.*

Пусть $P = P(t)$ - показатель размера колонии в момент времени $t$.

Составим обыкновенное дифференциальное уравнение:

$$ \frac{dP}{dt} = 0.7P =>  \frac{dP}{P} = 0.7dt$$

$$ \int\frac{dP}{P} = \int0.7dt$$

$$ \ln|P| = 0.7t + C $$

$$ P(t) = e^{0.7t + C} = Ce^{0.7t}$$

Предположим, что динамика роста колонии простейших описывается следующей моделью (модель Мальтуса):

$$P(t) = P_0e^{rt}$$

где $P_0 = P(0)$ — исходная численность колонии, $r$ — темп прироста колонии, $t$ — время.

Тогда модель для нашей задачи примет вид:

$$ P(t) = 2e^{0.7t} $$

```python
p = 2.0 # начальное значение биомассы
X = np.arange(0, 7)
y_dif_1 = [p] # значения модели, полученной с помощью диф. ур.
for t in range(1, len(X)):
    p = 2*exp(0.7*t)
    y_dif_1.append(p)
    print(f'Значение функции на {t} шаге: {round(p, 1)}')
```

Через 6 дней размер колонии будет составлять:

$$ P(6) = 2e^{0.7*6} = \textbf{133 особи} $$ 

Численное решение задачи примет следующий вид:

```python
# с шагом = 1
p = 2.0
X_1 = np.arange(0, 7) # задание расчетной сетки
y_raz_1 = [p] # значения модели, полученной с помощью разн. ур.
for t in range(1, len(X_1)):
    p = 1.7*p
    y_raz_1.append(p)
    print(f'Значение функции на {t} шаге: {round(p, 1)}')
```

$$ P(6) = \textbf{48 особей} $$ 

```python
# с шагом = 1/70
p = 2.0
h = 1/70 # шаг
X_2 = np.arange(0, 6+h, h)
y_raz_2 = [p] # значения модели, полученной с помощью разн. ур. с шагом 1/70
for t in range(1, len(X_2)):
    p = 1.01*p
    y_raz_2.append(p)
    if t == 420:
        print(f'Значение функции на {t} шаге: {round(p, 1)}')
```

$$ P(6) = \textbf{130 особей} $$ 

```python
plt.plot(X, y_dif_1, '-o', label='Обыкновенное диф. ур.')
plt.plot(X_1, y_raz_1, '-o', label='Разностное ур.')
plt.plot(X_2, y_raz_2, label='Разностное ур. с шагом 1/70')
plt.title('Динамика изменения численности колонии')
plt.xlabel('Шаги')
plt.ylabel('Численность колонии, шт. особей')
plt.legend(fontsize=8)
plt.show()
```

##### 2. Построить математические модели для решения задачи в виде: (а) обыкновенного дифференциального уравнения; (б) разностного уравнения; (в) разностного уравнения с половинным шагом. Решить численно задачу, пользуясь указанными моделями. Отобразить полученные графики изменения численности популяции (всего три графика). Объяснить разницу между решениями.

*Рассматривается популяция насекомых, биомасса которой измеряется в килограммах. Естественный прирост биомассы составляет 30% в год. Популяция также контролируется популяцией хищников, которая уничтожает 6 кг насекомых в год. (а) Найти модель, описывающую размер популяции P в любой момент времени t; (б) Определить численность популяции через 4 года, если начальная биомасса составляет 15 кг.*

Пусть $P = P(t)$ - показатель размера популяции в момент времени $t$.
#### а) Обыкновенное дифференциальное уравнение:

$$ \frac{dP}{dt} = 0.3P - 6 =>  \frac{dP}{P-20} = 0.3dt$$

$$ \int\frac{dP}{P-20} = \int0.3dt$$

$$ \ln|P-20| = 0.3t + C $$

$$ P(t) = e^{0.3t + C} + 20 = Ce^{0.3t} + 20$$

Так как по условию $P(0) = 15$, то $Ce^{0.3*0} + 20 = 15 => С = -5$

$$P(4) = -5e^{0.3*4} + 20 = \textbf{3.4 кг}$$

```python
p = 15.0 # начальное значение биомассы
X = np.arange(0, 5)
y_dif = [p] # значения модели, полученной с помощью диф. ур.
for t in range(1, len(X)):
    p = -5*exp(0.3*t)+20
    y_dif.append(p)
    print(f'Значение функции на {t} шаге: {round(p, 1)}')
```

#### б) Разностное уравнение:

$$P(t+1) = 1.3P(t) - 6$$

```python
p = 15.0
y_raz = [p] # значения модели, полученной с помощью разн. ур.
for t in range(1, len(X)):
    p = 1.3*p - 6
    y_raz.append(p)
    print(f'Значение функции на {t} шаге: {round(p, 1)}')
```

$$P(4) = \textbf{5.7 кг}$$

#### в) Разностное уравнение с половинным шагом:


$$P(t+0.5) = 1.15P(t) - 3$$

```python
h_half = 0.5 # шаг
p = 15.0
X_half = np.arange(0, 4+h_half, h_half)
y_half = [p] # значения модели, полученной с помощью разн. ур. с половинным шагом
for t in range(1, len(X_half)):
    p = 1.15*p - 3
    y_half.append(p)
    print(f'Значение функции на {t} шаге: {round(p, 1)}')
```

$$P(4) = \textbf{4.7 кг}$$

### Графики

```python
plt.plot(X, y_dif, '-o', label='Обыкновенное диф. ур.')
plt.plot(X, y_raz, '-o', label='Разностное ур.')
plt.plot(X_half, y_half, '-o', label='Разностное ур. с половинным шагом')
plt.title('Динамика изменения численности популяции')
plt.xlabel('Шаги')
plt.ylabel('Численность популяции, кг')
plt.legend(fontsize=8)
arrowprops={'arrowstyle': '-', 'ls':'--'}
plt.annotate(str(3.4), xy=(4, 3.4), xytext=(0, 3.4), 
             textcoords=plt.gca().get_yaxis_transform(),
             arrowprops=arrowprops,
             va='center', ha='right')
plt.annotate(str(5.7), xy=(4, 5.7), xytext=(0, 5.7), 
             textcoords=plt.gca().get_yaxis_transform(),
             arrowprops=arrowprops,
             va='center', ha='right')
plt.annotate(str(4.7), xy=(4, 4.7), xytext=(0, 4.7), 
             textcoords=plt.gca().get_yaxis_transform(),
             arrowprops=arrowprops,
             va='center', ha='right')
plt.show()
```

##### Эксперимент с шагом 1/30

```python
h_small = 1/30
p = 15.0
X_small = np.arange(0, 4+h_small, h_small)
y_small = [p]
for t in range(1, len(X_small)):
    p = 1.01*p - 0.2
    y_small.append(p)
    if t == 120:
        print(f'Значение функции на {t} шаге: {round(p, 1)}')
```

```python
plt.plot(X, y_dif, '-o', label='Обыкновенное диф. ур.')
plt.plot(X, y_raz, '-o', label='Разностное ур.')
plt.plot(X_half, y_half, '-o', label='Разностное ур. с половинным шагом')
plt.plot(X_small, y_small, label='Разностное ур. с шагом 1/30')
plt.xlabel('Шаги')
plt.ylabel('Численность популяции, кг')
plt.legend(fontsize=8)
plt.title('Динамика изменения численности популяции')
arrowprops={'arrowstyle': '-', 'ls':'--'}
plt.annotate(str(3.4), xy=(4, 3.4), xytext=(0, 3.4), 
             textcoords=plt.gca().get_yaxis_transform(),
             arrowprops=arrowprops,
             va='center', ha='right')
plt.annotate(str(5.7), xy=(4, 5.7), xytext=(0, 5.7), 
             textcoords=plt.gca().get_yaxis_transform(),
             arrowprops=arrowprops,
             va='center', ha='right')
plt.annotate(str(4.7), xy=(4, 4.7), xytext=(0, 4.7), 
             textcoords=plt.gca().get_yaxis_transform(),
             arrowprops=arrowprops,
             va='center', ha='right')
plt.annotate(str(3.5), xy=(4, 3.5), xytext=(0, 3.5), 
             textcoords=plt.gca().get_yaxis_transform(),
             arrowprops=arrowprops,
             va='center', ha='right')
plt.show()
```

### Вывод
Разностными уравнениями можно численно решать дифференциальные уравнения. При этом точность аппроксимации разностным уравнением зависит от размера шага, а погрешность имеет свойство накапливаться (в данном примере в последней точке погрешность принимает наибольшее значение). При этом нельзя сказать что модель, полученная при решении диф. уравнения, лучшим образом описывает исследуемый процесс, т.к. не учтены все факторы.
