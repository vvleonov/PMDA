```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
import warnings
warnings.filterwarnings("ignore")
```

```python
def dice_turn_unif(k):
    return np.random.uniform(1, 7, k)
```

```python
def dice_turn(k):
    return np.random.randint(1, 7, k)
```

```python
first_player = dice_turn(1000)
second_player = dice_turn(1000)
```

```python
fig, axs = plt.subplots(2)
axs[0].hist(first_player, bins=6)
axs[1].hist(second_player, color='red', bins=6)
plt.show()
```

```python
ufirst_player = dice_turn_unif(1000)
usecond_player = dice_turn_unif(1000)
```

```python
fig, axs = plt.subplots(2)
axs[0].hist(ufirst_player, bins=6)
axs[1].hist(usecond_player, color='red', bins=6)
plt.show()
```

```python
from collections import Counter
```

```python
observed = dice_turn(100000000)
```

```python
expected =[]
for i in range(6):
    expected.append(100000000/6)
```

```python
expected
```

```python
obs = list(Counter(observed).values())
```

```python
chi = 0
for i in range(6):
    chi += ((obs[i]-expected[i])**2)/expected[i]
```

```python
chi
```

```python
stats.chi2.ppf(q = 0.95, # Find the critical value for 95% confidence*
                      df = 5)
```

```python
1 - stats.chi2.cdf(x=chi, df=5)
```
